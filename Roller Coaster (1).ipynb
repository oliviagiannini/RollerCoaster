```python
import pandas as pd
import math as m
roller = pd.read_csv('rollercoaster.csv')
roller = roller.rename(columns={'Height (feet)': 'Height'})
roller.at[111, 'Height'] = 200
roller['Height'] = roller['Height'].astype(float)
tallest = roller['Height'].max()
shortest_height = roller['Height'].min()
# print(tallest)
# print(shortest_height)
# scale = (tallest + shortest_height)/11
# print(scale)


def height_score(height):
    if 29 <= height < 73:
        return 1
    if 73 <= height < 117:
        return 2
    if 117 <= height < 161:
        return 3
    if 161 <= height < 205:
        return 4
    if 205 <= height < 249:
        return 5
    if 249 <= height < 293:
        return 6
    if 293 <= height <= 337:
        return 7
    if 337 <= height < 381:
        return 8
    if 381 <= height < 425:
        return 9
    else:
        return 10


roller['height score'] = roller.Height.apply(height_score)

roller['Length (feet)'] = roller['Length (feet)'].astype(float)
longest_length = roller['Length (feet)'].max()
shortest_length = roller['Length (feet)'].min()
# print(longest_length)
# print(shortest_length)
# scale = (longest_length + shortest_length)/11
# print(scale)
roller['Length (feet)'] = roller['Length (feet)'].fillna(0)
roller['Length (feet)'] = roller['Length (feet)'].replace(' ', 0)
# roller.loc[53]['Length']
# def length_score(length):


def length_score(length):
    if 215 <= length < 974:
        return 1
    if 974 <= length < 1733:
        return 2
    if 1733 <= length < 2492:
        return 3
    if 2492 <= length < 3251:
        return 4
    if 3251 <= length < 4010:
        return 5
    if 4010 <= length < 4769:
        return 6
    if 4769 <= length <= 5528:
        return 7
    if 5528 <= length < 6287:
        return 8
    if 6287 <= length < 7046:
        return 9
    else:
        return 10


roller['length score'] = roller['Length (feet)'].apply(length_score)

most = roller['Number of Inversions'].max()
least = roller['Number of Inversions'].min()
roller.drop([300, 301, 302, 303, 304], inplace=True)
roller['Number of Inversions'] = roller['Number of Inversions'].fillna(0)
roller['Number of Inversions'] = roller['Number of Inversions'].astype(int)
# print(most)
# print(least)
# scale = (most + least)/11
# print(scale)


def inversion_number(inversion):
    if inversion == 0:
        return 1
    if 1 <= inversion < 3:
        return 2
    if 3 <= inversion < 5:
        return 3
    if 5 <= inversion < 7:
        return 4
    if 7 <= inversion < 9:
        return 5
    if 9 <= inversion < 10:
        return 6
    if 10 <= inversion < 11:
        return 7
    if 11 <= inversion < 12:
        return 8
    if 12 <= inversion < 13:
        return 9
    else:
        return 10


roller['inversion score'] = roller['Number of Inversions'].apply(inversion_number)

roller['Duration (min:sec)'] = roller['Duration (min:sec)'].fillna('0:00')
roller['Duration (min:sec)'] = roller['Duration (min:sec)'].replace(' ', '0:00')
# roller.loc[2]['Duration (min:sec)']


def convert(t):
    time = str(t).split(':')
    time_in_sec = int(time[0])*60+int(time[1])
    return time_in_sec


roller['Duration in sec'] = roller['Duration (min:sec)'].apply(convert)
longest_time = roller['Duration in sec'].min()
shortest_time = roller['Duration in sec'].max()
# print(longest_time)
# print(shortest_time)
# scale = (longest_time+shortest_time)/11
# print(scale)


def duration_score(time):
    if 0 <= time < 29.5:
        return 1
    if 59 <= time < 88.5:
        return 2
    if 88.5 <= time < 118:
        return 3
    if 118 <= time < 147.5:
        return 4
    if 147.5 <= time < 177:
        return 5
    if 177 <= time < 206.5:
        return 6
    if 206.5 <= time < 236:
        return 7
    if 236 <= time < 265.5:
        return 8
    if 265.5 <= time < 295:
        return 9
    else:
        return 10


roller['duration score'] = roller['Duration in sec'].apply(duration_score)


def Type(type_):
    if type_ == 'Sit Down':
        return 5
    if type_ == 'Inverted':
        return 8
    else:
        return 10


roller['type score'] = roller['Type'].apply(Type)

roller = roller.rename(columns={' Speed (mph)': 'Speed'})
roller['Speed'].fillna(0, inplace=True)
roller['Speed'] = roller['Speed'].astype(float)
roller['Speed'].replace(' ', 0, inplace=True)


def findspeed(x):
    if x['Speed'] == 0:
        speed = m.sqrt(2*7*x['Height'])
        return speed*2.237
    else:
        return x['Speed']


roller['speed'] = roller.apply(findspeed, axis=1)
fastest = roller['speed'].max()
slowest = roller['speed'].min()
# print(fastest)
# print(slowest)
# scale = (fastest-slowest)/11
# print(scale)


def speed_score(s):
    if 28 <= s < 39:
        return 1
    if 39 <= s < 50:
        return 2
    if 50 <= s < 61:
        return 3
    if 61 <= s < 72:
        return 4
    if 72 <= s < 83:
        return 5
    if 83 <= s < 94:
        return 6
    if 94 <= s < 105:
        return 7
    if 105 <= s < 116:
        return 8
    if 116 <= s < 127:
        return 9
    else:
        return 10


roller['speed score'] = roller['speed'].apply(speed_score)
roller['final score'] = roller['height score'] + roller['length score'] + roller['inversion score'] + roller['duration score'] + roller['type score'] + roller['speed score']
best = roller.nlargest(10, ['final score'])
print(best)
```
